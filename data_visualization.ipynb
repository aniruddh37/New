```python
# Impporing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.datasets import load_iris , load_wine
```

# Pandas Visualizations



## Pandas Plot 1: Line Plot (Time Series)

```python
data = {
    'months': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'sales': [200, 240, 300, 280, 350, 400] 
}
df = pd.DataFrame(data) 
df.plot(x='months', y='sales', kind='line', marker='o', title='Monthly Sales')
plt.show()
```

### Insights:
- The line plot shows clear upward trend in sales over the months.
- Sales increased steadily after march, indicating imporoved performance.
-  Minor fluctuation in April suggests possible seasonal effects or market changes.
- Overall, the trend is positive, indicating growth in sales over time.

## Pandas Plot 2: Bar Plot (Category Comparison)

```python
data = {
    'Department' : ['HR', 'IT','Finance','Marketing'],
    'Employees' : [15, 40, 25, 30]
}
df = pd.DataFrame(data)
df.plot(x='Department', y='Employees', kind='bar', title='Employees by Department')
plt.show()
```

### Insights:
- IT has the highest number of employees, indicating its critical role in the organization.
- HR has the least number of employees, which may suggest a lean structure in that department.
- Finance and Marketing have a moderate number of employees, reflecting balanced resource allocation.
- Bar plot makes categorical comparisons easy to visualize and understand.

## Pandas Plot 3: Histogram (Distribution)

```python
ages = [22, 25, 29, 35, 40, 28, 31, 45, 50, 23, 27]
df = pd.DataFrame(ages, columns=['Age'])
df.plot(kind='hist', bins=5, title='Age Distribution')
plt.show()
```

### Insights:
- The histogram shows that most individuals fall within the 25-35 age range.
- The distribution is right-skewed, indicating a smaller number of older individuals.
- Fewer older individuals act as outliers in this dataset.
- Histograms helps understand population spread and concentration. 

##   Pandas Plot 4: Box Plot (Outliers & Spread)

```python
scores = [55, 60, 65, 70, 72, 75, 78, 85, 90, 95]
df = pd.DataFrame({'Scores': scores})
df.plot(kind='box', title='Scores Distribution')
plt.show()
```

### Insights:
- The box plot indicates that the median score is around 75, showing central tendency.
- The interquartile range (IQR) suggests moderate variability in scores.
- There are no significant outliers, indicating consistent performance among individuals.
- Box plots are useful for visualizing data spread and identifying potential outliers.

## Pandas Plot 5: Scatter Plot (Relationship)

```python
data = {
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7],
    'Marks': [40, 45, 55, 65, 70, 78, 85]
}
df = pd.DataFrame(data)
df.plot(x='Hours_Studied', y='Marks', kind='scatter', title='Hours Studied vs Marks')
plt.show()
```

### Insights:
- Marks tend to increase with the number of hours studied, indicating a positive correlation.
- The relationship appears linear, suggesting consistent improvement with more study time.
- There are no apparent outliers, indicating that all data points follow the trend.
- Scatter plots are useful for detecting correlations between two continuous variables.

## Pandas Plot 6: Area Plot (Cumulative Trend)

```python
data = {
    'Year': [2021, 2022, 2023, 2024, 2025],
    'Revenue': [50, 60, 75, 90, 120]
    
}
df = pd.DataFrame(data)
df.plot(x='Year', y='Revenue', kind='area', title='Revenue Over Years')
plt.show()
```

### Insights:
- Revenue shows a consistent upward trend over the years, indicating business growth.
- The increase in revenue accelerates after 2023.
- The area plot emphasizes cumulative growth over time effectively.
- Area plots are useful for visualizing overall business expansion.

## Pandas Plot 7: Pie Chart (Proportion)

```python
data = {
    'Platform': ['Web', 'Mobile', 'Desktop'],
    'Users': [60, 30, 10 ]
}
df = pd.DataFrame(data)
df.plot(kind='pie', y='Users', labels=df['Platform'], autopct='%1.1f%%',
        title='User Platform Distribution', startangle=90)
plt.ylabel('')
plt.show()
```

### Insights:
- Majority of users access the platform via Web.
- Mobile users make up a significant portion, while Desktop users are minimal.
- The pie chart effectively illustrates the distribution of users across platforms.

## Pandas Plot 8: KDE Plot (Smooth Distribution)

```python
data = {
    'Income': [20, 25, 30, 35, 40, 45, 60, 80]
}
df = pd.DataFrame(data)
df.plot(kind='kde', title='Income Distribution (KDE)')
plt.show()
```

### Insights:
- Income distribution shows a peak around 40-50, indicating most individuals earn within this range.
- Distrubution shows a right skew, suggesting a smaller number of high-income earners.
- KDE plot provides a smooth estimate of the data distribution, making it easier to identify trends.

## Pandas Plot 9: Stacked Bar Plot

```python
data = {
    'Year': [2022, 2023, 2024, 2025],
    'Product_A': [30, 40, 50, 60],
    'Product_B': [20, 30, 45, 55]
}
df = pd.DataFrame(data)
df.plot(kind='bar', x='Year', stacked=True, title='Product Sales Over Years')
plt.show()
```

### Insights:
- Product_A consistently outperforms Product_B across all years.
- Both products show growth over the years, indicating increasing sales.
- Stacked bar help compare multiple categories over time effectively.

## Pandas Plot 10: Rolling Mean (Trend Smoothing)

```python
data = {
    'Sales': [200, 220, 250, 270, 260, 300, 320]
}
df = pd.DataFrame(data)
df['Rolling_Mean'] = df['Sales'].rolling(window=3).mean()
df.plot(y=['Sales', 'Rolling_Mean'], title='Sales with Rolling Mean')
plt.show()
```

### Insights:
- Rolling mean smooths out short-term fluctuations, revealing the underlying trend in sales.
- The trend indicates a steady increase in sales over time.
- Rolling mean is useful for identifying long-term trends in time series data.

# Matplotlib Visualizations

## Matplotlib Plot 1: Simple Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 40]
plt.plot(x, y, marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Insights:
- Performance increases steadily over time.
- The trend is generally upward, indicating growth.
- No major fluctuations are observed.
- Line plots are ideal for trend analysis.

## Matplotlib Plot 2: Histogram

```python
values = [5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(values, bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

### Insights:
- Values are evenly distributed across the bins.
- The histogram shows no significant skewness.
- Histogram is useful for visualizing frequency distribution of data.

## Matplotlib Plot 3: Bar Chart

```python
Categories = ['A', 'B', 'C', 'D']
Values = [12, 19, 7, 15]
plt.bar(Categories, Values)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### Insights:
- Category B has the highest value, indicating its prominence.
- Category C has the lowest value, suggesting it may need attention.
- Bar charts are effective for comparing categorical data.

## Matplotlib Plot 4: Scatter Plot

```python
x = [1, 2, 3, 4, 5, 6]
y = [5, 7, 8, 11, 14, 18]
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Insights:
- Output increases with increase in input.
- There is a positive correlation between the two variables.
- No extreme outliers are present.
- Scatter plots are useful for visualizing relationships between variables.

## Matplotlib Plot 5: Multiple Line Plot

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [8, 18, 28, 35, 45]
plt.plot(x, y1, marker='o', label='Product A')
plt.plot(x, y2, marker='o', label='Product B')
plt.title('Product performance Comparison')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

### Insights:
- Both products show an upward trend, indicating growth.
- Product A consistently outperforms Product B.
- The gap between the two products remains relatively constant.
- Multiple line enable comparative trend analysis.

## Matplotlib Plot 6: Horizontal Bar Chart

```python
Names = ['Alice', 'Bob', 'Charlie', 'Aniruddh']
Scores = [78, 85, 69, 90]
plt.barh(Names, Scores)
plt.title('Horizontal Bar Chart')
plt.xlabel('Scores')
plt.ylabel('Names')
plt.show()
```

### Insights:
- Aniruddh has the highest score, indicating strong performance.
- Charlie has the lowest score, suggesting room for improvement.
- Horizontal bar charts are useful for comparing categories with long names.

## Matplotlib Plot 7: Subplots

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [40, 30, 20, 10]
plt.subplot(1, 2, 1)
plt.plot(x, y1, marker='o', color='blue')
plt.title('Increasing Trend')
plt.subplot(1, 2, 2)
plt.plot(x, y2, marker='o', color='red')
plt.title('Decreasing Trend')
plt.show()
```

### Insights:
- First subplot shows an increasing trend, while the second shows a decreasing trend.
- Subplots allow for side-by-side comparison of different data trends.
- Useful for comparing multiple patterns simultaneously.

## Matplotlib Plot 8: Error Bar Plot

```python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]
error = [2, 3, 2, 4]
plt.errorbar(x, y, yerr=error, fmt='o')
plt.title('Error Bar Plot')
plt.xlabel('Experiment')
plt.ylabel('Value')
plt.show()
```

### Insights:
- Error margins vary across experiments, indicating differing levels of uncertainty.
- Experiment 4 has the highest value but also the largest error margin.
- Error bars are useful for visualizing variability in data measurements.

## Matplotlib Plot 9: Log Scale Plot

```python
x = [1, 10, 100, 1000]
y = [10, 100, 1000, 10000]
plt.loglog(x, y)
plt.title('Log-Log Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Insights:
- Linear relationship appears as a straight line on log-log scale.
- indicates exponetial growth pattern.
- Useful for visualizing data that spans several orders of magnitude.

## Matplotlib Plot 10: Bubble Chart

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
size = [100, 300, 500, 700]
plt.scatter(x, y, s=size, alpha=0.5)
plt.title('Bubble Chart')
plt.xlabel('X-value')    
plt.ylabel('Y-value')
plt.show()
```

### Insights:
- Bubble size represents an additional variable, adding depth to the scatter plot.
- Larger bubbles indicate higher magnitude.
- Clear upward trend is observed in the data points.
- Bubble charts enhance scatter plots by incorporating a third dimension.

# Seaborn Visualizations

## Seaborn Plot 1: Histogram

```python
df = sns.load_dataset('tips')
df.head()
```

```python

sns.histplot(df['total_bill'], bins=10, kde=True)
plt.title('Histogram of Total Bill with KDE')
plt.show()
```

### Insights:
- Most bills fall within the lower range.
- The distribution is right-skewed, indicating some high-value bills.
- Few high-value bills act as outliers in the dataset.
- Histogram shows spending behavior of customers effectively.

## Seaborn Plot 2: Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

### Insights:
- Weekends (Sat, Sun) tend to have higher total bills compared to weekdays.
- Saturday has the widest range of total bills, indicating more variability.
- Presence of outliers indicates variability in customer spending.
- Box plots are effective for comparing distributions across categories.

##  Seaborn Plot 3: Scatter Plot with Hue

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)
plt.title('Scatter Plot of Total Bill vs Tip by Gender')
plt.show()
```

### Insights:
- Tip generally increases with total bill amount.
- Males tend to give higher tips compared to females.
- Scatter plots with hue help visualize relationships while incorporating categorical distinctions.

## Seaborn Plot 4: Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=df)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

### Insights:
- Violin plots show distribution shape and density of total bills by day.
- Weekends have wider distributions, indicating more variability in spending.
- Saturday show higher variability compared to other days.
- Violin plots combine box plot and KDE for detailed distribution insights.

## Seaborn Plot 5: Count Plot

```python
sns.countplot(x='day', data=df)
plt.title('Count Plot of Bills by Day')
plt.show()
```

### Insights:
- On Saturday, there are the highest number of bills.
- Sunday also has a high number of bills, indicating busy days.
- Thursday has the lowest number of bills, suggesting it is a slower day.
- Count plots are useful for visualizing frequency of categorical data.

## Seaborn Plot 6: Bar Plot (Statistical)

```python
sns.barplot(x='day', y='tip', data=df)
plt.title('Average tip by Day')
plt.show()
```

### Insights:
- On Sunday, average tip amount is the highest.
- Average tips are higher on weekends compared to weekdays.
- Variation in average tips suggests different tipping behaviors across days.
- Bar plots display statistical summaries effectively.

## Seaborn Plot 7: Heatmap (Correlation)

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Insights:
- Total bill and tip have a strong positive correlation, indicating that higher bills tend to result in higher tips.
- Correlation between total bill and size is moderate, suggesting larger groups tend to have higher bills.
- Heatmaps are useful for visualizing relationships between multiple variables simultaneously.

## Seaborn Plot 8: Pair Plot

```python
sns.pairplot(df)
plt.show()
```

### Insights:
- Total bill and tip have a strong positive correlation.
- Total bill and tip are right-skewed distribution.
- Size distribution is skewed.
- Size variable is categorcial with discrete values.
- Pair plots are useful for visualizing relationships between multiple variables in a dataset.

## Seaborn Plot 9: Regression Plot

```python
sns.regplot(x='total_bill', y='tip', data=df)
plt.title('Regression Plot of Total Bill vs Tip')
plt.show()
```

### Insights:
- Regression line shows positive linear relationship.
- As total bill increases, tip amount also tends to increase.
- Linear plot is useful for understanding liner relationships between two continuous variables.

## Seaborn Plot 10: Categorical Plot

```python
sns.catplot(x='day', y='total_bill', kind='box', data=df)
plt.title('Categorical Plot of Total Bill by Day')
plt.show()
```

### Insights:
- Shows distribution of total bills across different days.
- Weekends tend to have higher total bills compared to weekdays.
- Box plots effectively display spread and outliers in total bill data by day.
- catplot allows flexible categorical data visualization.

## Plotly Plot 10: Animated Scatter Plot

```python
df = px.data.gapminder()

fig = px.scatter(
    df,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    color='continent',
    animation_frame='year',
    animation_group='country',
    log_x=True,
    title='GDP vs Life Expectancy Over Time'
)
fig.show()
```

### Insights:
- Global life expectancy improves over time.
- Economic growth correlates with longer life.
- Regional disparities persist.
- Animation reveals long-term trends clearly.
